```python
data = pd.read_csv('DS_Jobs.csv')
```

```python
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data.columns
```

```python
data.columns = ['title', 'salary', 'description', 'rating',
       'company', 'location', 'headquarters', 'size', 'founded',
       'type', 'industry', 'sector', 'revenue', 'competitors']
```

# Salary Parsing

```python
data.salary.value_counts()
```

```python
data = data[data['salary'] != '-1']
```

```python
data.shape
```

```python
data.salary.unique()
```

```python
salary = data.salary.apply(lambda x : x.split('(')[0])
```

```python
salary = salary.apply(lambda x : x.replace('K',''))
```

```python
salary = salary.apply(lambda x : x.replace('$',''))
```

```python
data['hourly'] = salary.apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
data['emp_provided'] = salary.apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
salary = salary.apply(lambda x: x.lower().replace('employer provided salary:',''))
```

```python
salary = salary.apply(lambda x: x.lower().replace('per hour',''))
```

```python
salary_l = salary.apply(lambda x: int(x.split('-')[0]))
salary_u = salary.apply(lambda x: int(x.split('-')[1]))
```

mean_salary = (salary_l + salary_u)/2

```python
data['mean_salary'] = mean_salary*1000
```

```python
data['mean_salary'] = data.apply(lambda x : x['mean_salary']*1920 if x['hourly']==1 else x['mean_salary'], axis = 1)
```

```python
data['mean_salary']
```

# Company name parsing

```python
data['company'] = data.apply(lambda x: x['company'] if x['rating'] == -1 else x['company'][:-3], axis = 1)
```

```python
data.company
```

# Splitting location to state and city

```python
data.location
```

```python
data['city'] = data['location'].apply(lambda x: x.split(',')[0])
data['state'] = data['location'].apply(lambda x: x.split(',')[1])
```

```python
data['job_at_hq'] = data.apply(lambda x: 1 if x.location == x.headquarters else 0, axis = 1)
```

# Company Age

```python
data['age'] = data.founded.apply(lambda x: x if x < 1 else 2020-x)
```

   # Categorizing Titles and Seniority

```python
def title_cat(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr'in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'junior' in title.lower():
        return 'junior'
    else:
        return 'na'
```

```python
data['title_cat'] = data['title'].apply(title_cat)
data.title_cat.value_counts()
```

```python
data['seniority'] = data['title'].apply(seniority)
data.seniority
```

```python
data.seniority.value_counts()
```

```python
# fix state Los Angeles
data['state'] = data['state'].apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
```

```python
data['desc_len'] = data['description'].apply(lambda x: len(x))
```

```python
data['company'] = data['company'].apply(lambda x : x.replace('\n',''))
```
